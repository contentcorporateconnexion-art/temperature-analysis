# **Importing LIB**

```python
import pandas as pd
import numpy as np
```

# **Create Sample Temperature Dataset**

```python
data = {
    "Day": ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"],
    "Temperature": [28, 30, 31, 29, 32, 33, 28]
}

df = pd.DataFrame(data)

print("Temperature Data:\n")
df

```

# **Convert Temperature column to NumPy array**

```python

temp_array = df["Temperature"].to_numpy()

```

# **Basic Temperature Analysis using NumPy**

```python

avg_temp = np.mean(temp_array)
max_temp = np.max(temp_array)
min_temp = np.min(temp_array)
std_temp = np.std(temp_array)

print("\n----- NUMPY ANALYSIS -----")
print("Average Temperature:", avg_temp)
print("Maximum Temperature:", max_temp)
print("Minimum Temperature:", min_temp)
print("Standard Deviation:", round(std_temp, 2))

```

# **Add Analysis Columns to DataFrame**

```python

df["Deviation"] = df["Temperature"] - avg_temp
df["Temp_Category"] = np.where(df["Temperature"] >= avg_temp, "High", "Low")

print("\nData with Analysis:\n")
df

```

# **Save the analyzed data**

```python
df.to_csv("temperature_analysis.csv", index=False)
print("\nCSV file saved: temperature_analysis.csv")
```
